```python
plt.bar(data.SepalLengthCm,data.SepalWidthCm,label = "Sepal",width = 0.5)
plt.bar(data.PetalLengthCm,data.PetalWidthCm,label = "Petal",width = 0.5)
plt.xlabel("width")
plt.ylabel("length")

plt.legend()  #indication of different color in the graph 
plt.title("Length and Breadth of sepal and petal")
```

```python
#histogram
```

```python

x = data["SepalLengthCm"]
  
plt.hist(x, bins = 20, color = "green")
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
```

```python
x = data.SepalWidthCm
  
plt.hist(x, bins = 20, color = "green")
plt.title("Sepal Width in cm")
plt.xlabel("Sepal_Width_cm")
plt.ylabel("Count")
```

```python
x = data.PetalLengthCm
  
plt.hist(x, bins = 20, color = "green")
plt.title("Petal Length in cm")
plt.xlabel("Petal_Length_cm")
plt.ylabel("Count")
  
```

```python
x = data.PetalWidthCm
  
plt.hist(x, bins = 20, color = "green")
plt.title("Petal Width in cm")
plt.xlabel("Petal_Width_cm")
plt.ylabel("Count")
```
